```
# load an example dataset
from vega_datasets import data
cars = data.cars()

import altair as alt

interval = alt.selection_interval()

points = alt.Chart(cars).mark_point().encode(
  x='Horsepower',
  y='Miles_per_Gallon',
  color=alt.condition(interval, 'Origin', alt.value('lightgray'))
).properties(
  selection=interval
)

histogram = alt.Chart(cars).mark_bar().encode(
  x='count()',
  y='Origin',
  color='Origin'
).transform_filter(interval)

points & histogram
```

```
from __future__ import print_function
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import statsmodels.api as sm
import plotly.graph_objs as go
```

```
loc_file = "tining.xlsx"
```

```
df = pd.read_excel(loc_file)
```

```
df.head()
vektabs = df['Vektendring_g'].abs()
dfabs = df
dfabs['Vektendring_g'] = df['Vektendring_g'].abs()
dfabs2 = dfabs
dfabs2['Vektendring_pst'] = dfabs2['Vektendring_pst'].abs()*100
dfabs2['Kjernetemp'] = dfabs2['Kjernetemp'].abs()
dfabs2.style.format("{:.2%}")
dfabs2
```

```
fig3 = px.scatter(dfabs2, x="Tid", y="Vektendring_pst", size="Kjernetemp", color="Tineprogram", labels={"Tid": "Tid (timer)", "Vektendring_pst": "Vektendring (%)", "Tineprogram": "Tineprogram"})
regline = sm.OLS(dfabs2['Vektendring_pst'], sm.add_constant(dfabs2['Tid'])).fit().fittedvalues
fig3.add_traces(go.Scatter(x=dfabs2['Tid'], y=regline, mode ="lines", marker_color="black",name="overall trendline"))
fig3.show()
```
